```python
import matplotlib.pyplot as plt        # conda install matplotlib
import numpy as np
```

# Gráfico de líneas

```python
x = np.linspace(0,51)
y= np.linspace(0,201)

plt.rcParams['axes.facecolor'] = 'lightgrey'
plt.plot(x,y)

plt.grid()
```

# Título y Etiquetas

```python
x = np.linspace(0,51)
y= np.linspace(0,201)

plt.title('Aprendiendo Matplotlib - 1ra Parte', fontdict={'fontsize':20})
plt.xlabel('Valores Eje X')
plt.ylabel('Valores Eje Y')

plt.plot(x, y)
plt.grid()
```

# Especificar los valores de los ejes

```python
x = np.linspace(0,51)
y= np.linspace(0,201)

plt.title('Matplotlib - 1ra Parte', fontdict={'fontsize':20})
plt.xlabel('Valores Eje X')
plt.ylabel('Valores Eje Y')

plt.xticks([5,15,25,35,45])
plt.yticks([50,100,150,200])

plt.plot(x, y)
plt.grid()
```

# Editar la línea

```python
x = np.linspace(0,51)
y= np.linspace(0,201)

plt.figure(figsize=(15,5))

# plt.plot(x, y, color='red',
#          linewidth=10,
#          linestyle='--',
#          marker='v',
#          markersize=20,
#          markeredgecolor='black')

plt.plot(x, y, color='green',
         linewidth=5,
         linestyle='-.',
         marker='o',
         markersize=15,
         markeredgecolor='red')

plt.title('Matplotlib - 1ra Parte', fontdict={'fontsize':20})
plt.xlabel('Valores Eje X')
plt.ylabel('Valores Eje Y')

plt.xticks([5,10,15,20,25,30,35,40,45,50])
plt.yticks([50,100,150,200])

plt.plot(x, y)
plt.grid()
```

# Agregar una segunda línea

```python
x = np.arange(0,31)
y= x*5
y2= x*2
plt.figure(figsize=(15,5))

plt.plot(x, y, color='red',
         linewidth=5,
         linestyle='--',
         marker='v',
         markersize=20,
         markeredgecolor='black')
plt.plot(x, y2, color='green')

plt.title('Matplotlib - 1ra Parte', fontdict={'fontsize':20})
plt.xlabel('Valores Eje X')
plt.ylabel('Valores Eje Y')

plt.xticks([5,10,15,20,25,30])


plt.plot(x, y)
plt.grid()
```

# Guardar el gráfico

```python
x = np.arange(1,31)
y= x*5
y2= x*2
plt.figure(figsize=(15,5))

plt.plot(x, y, color='purple', linewidth=5, linestyle='--', marker='v', markersize=20, markeredgecolor='black')
plt.plot(x, y2, color='green')

plt.title('Aprendiendo Matplotlib', fontdict={'fontsize':15})
plt.xlabel('Valores Eje X')
plt.ylabel('Valores Eje Y')

plt.xticks([5,10,15,20,25,30])

plt.savefig('Python con Héctor.png')

plt.grid()
```

# Scatter Plot

# Gráfico de Barras

```python
etiquetas = ['Alemania','Argentina','Brasil',
             'España','Francia','Inglaterra',
             'Italia','Portugal','Uruguay']
valores = [4,2,5,1,2,1,4,1,2]

plt.figure(figsize=(10,5))
plt.bar(etiquetas, valores)
plt.show()
```

# Agregar una línea con el valor promedio

```python
etiquetas = ['Alemania','Argentina','Brasil',
             'España','Francia','Inglaterra',
             'Italia','Portugal','Uruguay']
valores = [4,2,5,1,2,1,4,1,2]

plt.figure(figsize=(10,5))
grafico_barras = plt.bar(etiquetas, valores, color='aqua')


promedio = np.mean(valores)
plt.axhline(promedio, color='darkblue')

plt.show()
```

# Agregar símbolos en las barras

```python
etiquetas = ['Alemania','Argentina','Brasil',
             'España','Francia','Inglaterra',
             'Italia','Uruguay']
valores = [4,2,5,1,2,1,4,2]

plt.figure(figsize=(10,5))
grafico_barras = plt.bar(etiquetas, valores, color='goldenrod')

grafico_barras[0].set_hatch('/')
grafico_barras[1].set_hatch('*')
grafico_barras[2].set_hatch('-')
grafico_barras[3].set_hatch('.')
grafico_barras[4].set_hatch('|')
grafico_barras[5].set_hatch('+')
grafico_barras[6].set_hatch('\\')
grafico_barras[7].set_hatch('-.')


plt.show()
```
