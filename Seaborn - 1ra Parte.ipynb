```python
import pandas as pd
import seaborn as sns # pip install seaborn
import numpy as np
import matplotlib.pyplot as plt
```

```python
vendedores_df = pd.read_excel('datos de vendedores.xlsx')
vendedores_df.head(10)
```

# Barplot

```python
# Agrupamiento de regiones segÃºn cantidad de vendedores con ventas registradas
ventas_x_region = vendedores_df.groupby('REGION ASIGNADA')['VENTAS'].count()
ventas_x_region.head(7)
```

```python
sns.barplot(x = ventas_x_region.index,
            y = ventas_x_region)
```

```python
# Cambiamos el estilo y agregamos palette
sns.set_style("darkgrid")
sns.barplot(x = ventas_x_region.index,
            y = ventas_x_region,
            palette = 'Reds'
            )
```

```python
# Agreamos la lÃ­nea para el promedio
sns.set_style('darkgrid')
sns.barplot(x=ventas_x_region.index,
                y= ventas_x_region,
           palette= 'cool')

promedio_region = np.mean(ventas_x_region)
plt.axhline(promedio_region, color='black')
```

# Scatter

```python
sns.set_style("whitegrid")
sns.scatterplot(x = 'REGION ASIGNADA',
                y ='VENTAS',
                data = vendedores_df)
```

```python
# Cambiamos el estilo y el tamaÃ±o de los puntos.
# TambiÃ©n agregamos el argumento hue
sns.set_style("darkgrid")
grafico = sns.scatterplot(x = 'REGION ASIGNADA',
                          y = 'VENTAS',
                          data = vendedores_df,
                          hue ='CALIFICACION DADA POR LOS CLIENTES',
                          s = 80)
sns.move_legend(grafico, "upper left", bbox_to_anchor=(1, 1))
```

# Caja

# Violin

```python
sns.set_style("darkgrid")
sns.violinplot( x = 'REGION ASIGNADA',
                y = 'VENTAS',
                data = vendedores_df)
```

```python
sns.set_style("darkgrid")
sns.violinplot(x = 'REGION ASIGNADA',
               y = 'VENTAS',
               data = vendedores_df)
sns.swarmplot(x = 'REGION ASIGNADA',
              y = 'VENTAS',
              data = vendedores_df,
              color = "k",
              alpha = 0.8)
```
