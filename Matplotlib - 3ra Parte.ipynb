```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Precio de cierre ajustado desde 02-01-2019 hasta 02-12-2022
AAPL_df = pd.read_excel('AAPL.xlsx')
MSFT_df = pd.read_excel('MSFT.xlsx')
GOOG_df = pd.read_excel('GOOG.xlsx')
AMZN_df = pd.read_excel('AMZN.xlsx')
```

```python
AAPL_df.head()
```

```python
AAPL_df.set_index('Date', inplace=True)
MSFT_df.set_index('Date', inplace=True)
GOOG_df.set_index('Date', inplace=True)
AMZN_df.set_index('Date', inplace=True)
```

```python
AAPL_df.tail()
```

```python
fig, ax = plt.subplots(nrows = 4, ncols=1, figsize=(10,5), sharex=True)

ax[0].plot(AAPL_df.index, AAPL_df['Adj Close'])
ax[1].plot(MSFT_df.index, MSFT_df['Adj Close'])
ax[2].plot(GOOG_df.index, GOOG_df['Adj Close'])
ax[3].plot(AMZN_df.index, AMZN_df['Adj Close'])
```

```python
fig, ax = plt.subplots(nrows = 4, ncols=1, figsize=(18,11), sharex=True)

fig.set_facecolor('lightgrey')
fig.autofmt_xdate(rotation=30)

ax[0].plot(AAPL_df.index, AAPL_df['Adj Close'], label = "Apple", color='lime')
ax[1].plot(MSFT_df.index, MSFT_df['Adj Close'], label = "Microsoft", color='goldenrod')
ax[2].plot(GOOG_df.index, GOOG_df['Adj Close'], label = "Alphabet", color='orangered')
ax[3].plot(AMZN_df.index, AMZN_df['Adj Close'], label = "Amazon", color='magenta')

ax[0].set_title("\nComparaciÃ³n entre AAPL - MSFT - GOOG - AMZN\n", fontsize=22)
ax[3].set_xlabel("\nAÃ±o\n", fontsize=18)

ax[0].legend()
ax[1].legend()
ax[2].legend()
ax[3].legend()


ax[0].axvline(x=18310, color = 'black', linestyle='dashdot')
ax[1].axvline(x=18308, color = 'black', linestyle='dashdot')
ax[2].axvline(x=18312, color = 'black', linestyle='dashdot')
ax[3].axvline(x=18308, color = 'black', linestyle='dashdot')

ax[0].axvline(x=18340, color = 'black', linestyle=':')
ax[1].axvline(x=18340, color = 'black', linestyle=':')
ax[2].axvline(x=18340, color = 'black', linestyle=':')
ax[3].axvline(x=18335, color = 'black', linestyle=':')
```
